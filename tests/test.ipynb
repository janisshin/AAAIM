```python
import tellurium as te
import numpy as np

# Define the model with initial conditions for all species
r = te.loada ('''
    A -> B; k1*A
    B + C -> D; k2*B*C
    D -> E + A; k3*D
    E -> C; k4*E
    k1 = 0.1; A = 5; B = 0; C = 0; D = 0; E = 1
    k2 = 0.56; k3 = 0.45; k4 = 0.6
''')

# Simulate the model
m = r.simulate (0, 200, 40)

# Plot the results
r.plot()

```
